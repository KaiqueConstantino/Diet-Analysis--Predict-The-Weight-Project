```python
"""  
Questions to be ansswered:

1 - What is the age distribution of participants?

2 - What is the relationship between daily calorie intake and deficit/excess calories?

3 - What is the distribution of physical activity level and stress level?

4 - What is the relationship between study duration and weight change?

"""

# Importing the libraries
import pandas as pd  

# Importing the dataset
df = pd.read_csv('weight_change_dataset.csv')
```

1 - What is the age distribution of participants?

```python
import matplotlib.pyplot as plt

plt.hist(df['Age'], bins=10, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

2 - What is the relationship between daily calorie intake and deficit/excess calories?

```python
import seaborn as sns 
sns.scatterplot(x='Daily Calories Consumed', y='Daily Caloric Surplus/Deficit', data=df)
plt.xlabel('Daily Calories Consumed')
plt.ylabel('Daily Caloric Surplus/Deficit')
plt.title('Relationship between Daily Calories Consumed and Daily Caloric Surplus/Deficit')
plt.show()
```

3 - What is the distribution of physical activity level and stress level?

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.bar(df['Physical Activity Level'].value_counts().index, df['Physical Activity Level'].value_counts())
plt.xlabel('Physical Activity Level')
plt.ylabel('Frequency')
plt.title('Physical Activity Level Distribution')

plt.subplot(1, 2, 2)
plt.bar(df['Stress Level'].value_counts().index, df['Stress Level'].value_counts())
plt.xlabel('Stress Level')
plt.ylabel('Frequency')
plt.title('Stress Level Distribution')

plt.tight_layout()
plt.show()
```

4 - What is the relationship between study duration and weight change?

```python
sns.scatterplot(x='Duration (weeks)', y='Weight Change (lbs)', data=df)
plt.xlabel('Duration (weeks)')
plt.ylabel('Weight Change (lbs)')
plt.title('Relationship between Study Duration and Weight Change')
plt.show()
```
